````python
import requests
import pandas as pd
url_neo='https://lgbta.wikia.org/wiki/Neopronouns'
html = requests.get(url_neo).content
df_list = pd.read_html(html)
df = df_list[-1]
df
````

````python
neo_list=[]
dict_df={}
for df in df_list[:12]:
  print('---------------')
  print(df.shape,df)
  print('---------------')
  pr_=df.Pronoun[0]
  neo_list.append(pr_)
  dict_df[pr_]=df
case_list=list(df.Case.values)

neo_list,case_list
````

````python
# Initialize the dataframes

df_pronoun=pd.DataFrame(index=neo_list,columns=case_list)
df_pronounce=pd.DataFrame(index=neo_list,columns=case_list)
df_example=pd.DataFrame(index=neo_list,columns=case_list)
df_example
````

````python
for neo_pro in neo_list:
  df_neo=dict_df[neo_pro]
  for case in case_list:
    df_pronoun.loc[neo_pro,case]=df_neo.loc[df_neo.Case==case,'Pronoun'].values[0]
    df_example.loc[neo_pro,case]=df_neo.loc[df_neo.Case==case,'Example'].values[0]
    df_pronounce.loc[neo_pro,case]=df_neo.loc[df_neo.Case==case,'Pronunciation'].values[0]
````

````python
df_pronoun
````

````python
df_pronoun.loc['She']=['She','Her','Her','Hers','Herself']
df_pronoun.loc['He']=['He','Him','His','His','Himself']
df_pronoun.loc['They']=['They','Them','Their','Theirs','Themselves']
# Source: https://intercultural.uncg.edu/wp-content/uploads/Neopronouns-Explained-UNCG-Intercultural-Engagement.pdf
df_pronoun.loc['Ne']=['Ne','Nem','Nir','Nirs','Nemself']
df_pronoun
````
